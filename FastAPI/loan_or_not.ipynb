### 4) Variable selection and feature engineering

```python
# Selecting variables for the prediction
X = df[["Married","ApplicantIncome","Education","LoanAmount","Credit_History"]]
```

```python
X
```

```python
y = df[["Loan_Status"]]
```

```python
y
```

```python
# For "y" change text "N" and "Y" to boolean
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
y["Loan_Status"] = labelencoder.fit_transform(y["Loan_Status"])
```

```python
y
```

```python
# "Y" is 1 and "N" is 0, we check with labelencoder
labelencoder.classes_
```

```python
# With "X"
X["Married"] = labelencoder.fit_transform(X["Married"])
```

```python
X
```

```python
labelencoder.classes_
```

```python
X["Education"] = labelencoder.fit_transform(X["Education"])
```

```python
X
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

### IV) Building 3 prediction models

```python
# To import scaler to FastAPI, we need joblib
import joblib 
joblib.dump(scaler, "Scaler.pkl")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(predictions) :
    print("Accuracy score in model is {}".format(accuracy_score(y_test, predictions)))
```

1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
# Prediction
log_model.predict(X_test)
```

```python
log_predictions = log_model.predict(X_test)
```

```python
modelperformance(log_predictions)
```

2. K-Nearest Neighbors

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors" : [3,5,7,8,6],
              "weights" : ["uniform","distance"]}
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid)
gridkn.fit(X_train, y_train)
```

```python
gridkn.best_params_
```

```python
knpreds = gridkn.predict(X_test)
```

```python
modelperformance(knpreds)
```

3. Suport Vector Machine

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
param_grid_svc = {"C": [0.01,0.1,0.5], "kernel": ["linear", "rbf", "poly"]}
```

```python
gridsvc = GridSearchCV(svm, param_grid_svc)
```

```python
gridsvc.fit(X_train, y_train)
```

```python
gridsvc.best_params_
```

```python
svc_pred = gridsvc.predict(X_test)
```

```python
modelperformance(svc_pred)
```

### V) Deploy the model of machine learning on FastAPI

```python
joblib.dump(gridsvc, "model.pkl")
```
